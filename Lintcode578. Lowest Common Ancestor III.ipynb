# Lintcode578. Lowest Common Ancestor III

```python
"""
Definition of TreeNode:
class TreeNode:
    def __init__(self, val):
        this.val = val
        this.left, this.right = None, None
"""


class Solution:
    """
    @param: root: The root of the binary tree.
    @param: A: A TreeNode
    @param: B: A TreeNode
    @return: Return the LCA of the two nodes.
    """
    def lowestCommonAncestor3(self, root, A, B):
        # write your code here
        if not root or not A or not B:
            return None
            
        self.foundA = False
        self.foundB = False
        
        lca = self.findLCA(root,A,B)
        
        if self.foundA and self.foundB:
            return lca
        return None
        
    def findLCA(self,node,A,B):
        if not node:
            return None
        left_lca = self.findLCA(node.left,A,B)
        right_lca = self.findLCA(node.right,A,B)
        
        if node is A:
            self.foundA = True
        if node is B:
            self.foundB = True
            
        if node in (A,B) or (left_lca and right_lca):
            return node
        return left_lca or right_lca or None
        
```
